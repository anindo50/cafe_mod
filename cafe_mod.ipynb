```python
from urllib.request import urlopen as ureq
from bs4 import BeautifulSoup as soup
import pandas as pd
myurl = 'https://www.yellowpages.com/chicago-il/mip/cafe-il-cortile-8731718#reviews'
open = ureq(myurl)
pagehtml = open.read()
open.close()
pagesoup = soup(pagehtml, "lxml")

c_name = []
c_n = pagesoup.h1.text
c_name.append(c_n)
```

```python
item = []
for i in pagesoup.find_all(class_='title-price'):
    x = i.text.replace(u'\xa0', u' ')
    item.append(x)
```

```python
rate = []
r = pagesoup.find(class_='count').text
rate.append(r)
```

```python
ocr = []
for i in pagesoup.find_all(class_='cross-links'):
    h = i.text
    ocr.append(h)
```

```python
al = []

info = pagesoup.find_all('dl')

for re in info[0].find_all('dt'):
    al.append(re.text.replace(u'\xa0', u' '))
bl = []
for i in info[0].find_all('dd'):
    bl.append(i.text.replace(u'\xa0', u' '))
al_bl = []
for x,y in zip(al,bl):
    al_bl.append({'info tittle':x,'detail info':y})
```

```python
sm = []
for i in pagesoup.find_all(class_='smart-article'):
    h = i.text
sm.append(h)
```

```python
ca = pd.DataFrame({'cafe name':c_name})
ab = pd.DataFrame (al_bl)
mn = pd.DataFrame({'menu':item})
rs = pd.DataFrame({'cafe rating':rate})
oc = pd.DataFrame({'other cafe and restaurants':ocr})
sms = pd.DataFrame({'smart articals':sm})
df = pd.concat([ca,ab,mn,rs,oc,sms],axis=1)
df.to_csv('cafe_mod.csv',encoding = 'utf-8',header=True)
```
